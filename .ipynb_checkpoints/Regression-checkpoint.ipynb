### Plotting all the points in the dataset

```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
x = []
y = []

file_name = "C:/Users/DELL/Desktop/HITK ML NOTEBOOKS/MV_Reg_noisy_data.csv"
with open(file_name, mode = 'r') as file:
    csvFile = csv.reader(file)
    flag = 0
    for row in csvFile:
        if flag == 0:
            flag = 1
            continue
        else:
            x.append(float(row[0]))
            y.append(float(row[1]))

plt.scatter(x, y)
plt.show()
```

#### Where Î± is the learning parameter.

# PART 1: Observing how the Squared Sum Error decreases with each iteration for different hypotheses used   [training data = 70%, test data = 30%]  

## ***SIMPLE LINEAR REGRESSION

### Hypothesis function                                 y = w0 + w1*x

```python
alpha = 0.001          # Learning rate 

def lin_regression(num_of_iterations, num_of_training_samples):
    in_sample_error = []
    out_sample_error = []
    iteration = []
    i = 1
    w = [2, 2]         # Initial assumption
    while (i <= num_of_iterations):
        j = 0
        while j < num_of_training_samples:
            X = x[j]
            Y = y[j]
            w[0] -= alpha * ( w[0] + w[1]*X - Y)
            w[1] -= alpha * ( w[0] + w[1]*X - Y)*X
            j += 1
        
        j = 0
        in_error = 0
        while j < num_of_training_samples:
            X = x[j]
            Y = y[j]
            in_error += ((Y - w[0] - w[1]*X)**2)
            j += 1
        in_error = in_error / (2 * num_of_training_samples) 
        
        j = num_of_training_samples
        out_error = 0
        while j < 300:
            X = x[j]
            Y = y[j]
            out_error += ((Y - w[0] - w[1]*X)**2)
            j += 1
        out_error = out_error / (2 * (300 - num_of_training_samples))
        
        in_sample_error.append(in_error)
        out_sample_error.append(out_error)
        iteration.append(i)
        i +=1
    return w, iteration, in_sample_error, out_sample_error
```

### Plotting the gradual decrease of SSE  

```python
coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 210)
f = plt.figure()
f.set_figwidth(10)    
f.set_figheight(5)
plt.plot(iteration, in_sample_error, label = "in_sample error")
plt.plot(iteration, out_sample_error, label = "out_sample error")
plt.title("How Sum of Square Error(S.S.E.) decreases with each iteration....")
plt.xlabel("Iteration count --->")
plt.ylabel("S.S.E. --->")
plt.legend()
plt.show() 
```

###  Plotting the best fit curve  [y = w0 + w1*x]

```python
line_x = [0, 5]
line_y = [coefficients[0], 5*coefficients[1] + coefficients[0]]
plt.plot(line_x, line_y, color = 'red')
plt.scatter(x, y)
plt.show()
```

##  ***MULTIVARIATE REGRESSION [ degree = 2 (QUADRATIC) ]

### Hypothesis function y = w0 + w1*x + w2*x^2

```python
alpha = 0.001          # Learning rate 

def quad_regression(num_of_iterations, num_of_training_samples):
    in_sample_error = []
    out_sample_error = []
    iteration = []
    i = 1
    w = [2, 2, 2]      # Initial assumption
    while (i <= num_of_iterations):
        j = 0
        while j < num_of_training_samples:
            X = x[j]
            Y = y[j]
            w[0] -= alpha * ( w[0] + w[1]*X + w[2]*X*X - Y)
            w[1] -= alpha * ( w[0] + w[1]*X + w[2]*X*X - Y)*X
            w[2] -= alpha * ( w[0] + w[1]*X + w[2]*X*X - Y)*X**2
            j += 1
        
        j = 0
        in_error = 0
        while j < num_of_training_samples:
            X = x[j]
            Y = y[j]
            in_error += ((Y - w[0] - w[1]*X - w[2]*X*X)**2)
            j += 1
        in_error = in_error / (2 * num_of_training_samples) 
        
        j = num_of_training_samples
        out_error = 0
        while j < 300:
            X = x[j]
            Y = y[j]
            out_error += ((Y - w[0] - w[1]*X - w[2]*X*X)**2)
            j += 1
        out_error = out_error / (2 * (300 - num_of_training_samples))
        
        in_sample_error.append(in_error)
        out_sample_error.append(out_error)
        iteration.append(i)
        i +=1
    return w, iteration, in_sample_error, out_sample_error
```

### Plotting the gradual decrease of SSE

```python
coefficients, iteration, in_sample_error, out_sample_error = quad_regression(100, 210)
f = plt.figure()
f.set_figwidth(10)    
f.set_figheight(5)
plt.plot(iteration, in_sample_error, label = "in_sample error")
plt.plot(iteration, out_sample_error, label = "out_sample error")
plt.title("How Sum of Square Error(S.S.E.) decreases with each iteration....")
plt.xlabel("Iteration count --->")
plt.ylabel("S.S.E. --->")
plt.legend()
plt.show() 
```

### Plotting the best fit curve [y = w0 + w1*x + w2x^2]

```python
line_x = np.linspace(-1,6,5000)
line_y = coefficients[0] + coefficients[1]*line_x + coefficients[2]*line_x**2
plt.plot(line_x, line_y, color = 'red')
plt.scatter(x, y)
plt.show()
```

## ***MULTIVARIATE REGRESSION [ degree = 3 (CUBIC) ]

### Hypothesis function y = w0 + w1*x + w2*x^2 + w3*x^3

# PART 2: Observing how the results change with the increase in the number training samples for different hypotheses used

## SIMPLE LINEAR REGRESSION

```python
coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 5)
in_avg5 = sum(in_sample_error) / len(in_sample_error)
out_avg5 = sum(out_sample_error) / len(out_sample_error)

coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 10)
in_avg10 = sum(in_sample_error) / len(in_sample_error)
out_avg10 = sum(out_sample_error) / len(out_sample_error)

coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 20)
in_avg20 = sum(in_sample_error) / len(in_sample_error)
out_avg20 = sum(out_sample_error) / len(out_sample_error)

coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 50)
in_avg50 = sum(in_sample_error) / len(in_sample_error)
out_avg50 = sum(out_sample_error) / len(out_sample_error)

coefficients, iteration, in_sample_error, out_sample_error = lin_regression(100, 100)
in_avg100 = sum(in_sample_error) / len(in_sample_error)
out_avg100 = sum(out_sample_error) / len(out_sample_error)

in_y = [in_avg5, in_avg10, in_avg20, in_avg50, in_avg100]
out_y = [out_avg5, out_avg10, out_avg20, out_avg50, out_avg100]
range_x = [5, 10, 20, 50, 100]

f = plt.figure()
f.set_figwidth(10)    
f.set_figheight(5)
plt.plot(range_x, in_y, color ='maroon', label = "in_sample error")
plt.plot(range_x, out_y, color ='blue', label = "out_sample error")
plt.title("How Sum of Square Error(S.S.E.) decreases with increase in training set size (LINEAR REGRESSION) ....")
plt.xlabel("Training set size --->")
plt.ylabel("S.S.E. --->")
plt.legend()
plt.show()
```
