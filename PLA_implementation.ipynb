## Perceptron and graph implementation function

```python
import matplotlib.pyplot as plt
import numpy as np
import csv

threshold = 0
learning_rate = 0.001

def signum(weighted_sum):
    if weighted_sum >= threshold:
        return 1
    else:
        return -1
    
    
    
def perceptron(dataset,num_of_iterations):
    weight = [0.5, 0.5]
    error_count = []
    iteration = []
    i = 1
    while (i <= num_of_iterations):
        count = 0
        with open(dataset, mode = 'r') as file:
            csvFile = csv.reader(file)
            flag = 0
            for row in csvFile:
                if flag == 0:
                    flag = 1
                    continue
                x1 = float(row[0])
                x2 = float(row[1])
                w1 = weight[0]
                w2 = weight[1]
                y = float(row[2])
                if (signum(x1*w1 + x2*w2) == -1) and (y == 1):
                    count +=1
                    weight = [(weight[0] + x1*learning_rate) , (weight[1] + x2*learning_rate)]
                elif (signum(x1*w1 + x2*w2) == 1) and (y == -1):
                    count +=1
                    weight = [(weight[0] - x1*learning_rate) , (weight[1] - x2*learning_rate)]
        iteration.append(i)
        error_count.append(count)
        i +=1
    print("ITERATION ERROR_COUNT")
    for i in range(len(iteration)):
        print("\t"+str(iteration[i])+"\t"+str(error_count[i]))
    f = plt.figure()
    f.set_figwidth(10)
    f.set_figheight(3)
    plt.plot(iteration, error_count)
    plt.show() 
    
    
```

## First Dataset

```python
perceptron("PLA_Data1.csv", 50)
```
